```python
import pandas as pd
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np

# 读取原始网络数据
df = pd.read_csv(r'D:\metabolic network\转录组\Wgcna\GAPR\block_ppi_GAPR.txt', sep='\t')
G_original = nx.from_pandas_edgelist(df, '#Protein A', 'Protein B')

# 计算原始网络的割点
original_APs = set(nx.articulation_points(G_original))
original_nodes = set(G_original.nodes())
num_original_APs = len(original_APs)

# 定义要测试的移除比例，从1%到50%
percentages = np.arange(1, 51)  # 1% 到 50%

# 存储结果的列表
proportions_of_APs_remaining = []
proportions_of_APs_still_APs = []

for pct in percentages:
    # 计算需要移除的节点数量
    num_nodes_to_remove = int(len(G_original) * pct / 100)

    # 确保不会移除超过现有节点数量
    num_nodes_to_remove = min(num_nodes_to_remove, len(G_original))

    # 随机选择要移除的节点
    nodes_to_remove = random.sample(original_nodes, num_nodes_to_remove)

    # 创建网络副本并移除节点
    G_removed = G_original.copy()
    G_removed.remove_nodes_from(nodes_to_remove)

    # 更新剩余的原始割点（那些未被移除的割点）
    remaining_original_APs = original_APs - set(nodes_to_remove)
    num_remaining_original_APs = len(remaining_original_APs)

    if num_remaining_original_APs == 0:
        # 如果所有原始割点都被移除了
        proportions_of_APs_remaining.append(0)
        proportions_of_APs_still_APs.append(0)
        continue

    # 计算移除节点后的新割点
    new_APs = set(nx.articulation_points(G_removed))

    # 计算剩余的原始割点中仍然是割点的比例
    num_APs_still_APs = len(remaining_original_APs & new_APs)
    proportion_APs_still_APs = num_APs_still_APs / num_remaining_original_APs

    # 保存结果
    proportions_of_APs_remaining.append(num_remaining_original_APs / num_original_APs)
    proportions_of_APs_still_APs.append(proportion_APs_still_APs)

# 将结果保存到DataFrame中
df_results = pd.DataFrame({
    'Percentage of Nodes Removed (%)': percentages,
    'Proportion of Original APs Remaining': proportions_of_APs_remaining,
    'Proportion of Remaining APs Still APs': proportions_of_APs_still_APs
})

# 输出结果
print(df_results)

# 绘制结果
plt.figure(figsize=(12, 6))
plt.plot(percentages, proportions_of_APs_remaining, label='Proportion of Original APs Remaining')
plt.plot(percentages, proportions_of_APs_still_APs, label='Proportion of Remaining APs Still APs')
plt.xlabel('Percentage of Nodes Removed (%)')
plt.ylabel('Proportion')
plt.title('Network Robustness Analysis')
plt.legend()
plt.grid(True)
plt.show()

# 保存结果到Excel文件
df_results.to_excel(r'D:\文章\RORgt\network_robustness_analysis.xlsx', index=False)
```
